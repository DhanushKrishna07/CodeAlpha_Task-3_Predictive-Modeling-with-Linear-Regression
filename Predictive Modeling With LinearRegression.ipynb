## Checking if there any missing values

```python
df.isna().sum()
```

```python
# Filling missing values with the mean or median
df['crim'].fillna(df['crim'].median(), inplace=True)
df['zn'].fillna(df['zn'].median(), inplace=True)
df['indus'].fillna(df['indus'].mean(), inplace=True)
df['age'].fillna(df['age'].median(), inplace=True)
df['lstat'].fillna(df['lstat'].median(), inplace=True)

# Filling missing values with the mode
df['chas'].fillna(df['chas'].mode()[0], inplace=True)

# Checking the filled data
print(df.isna().sum())

# Displaying the DataFrame after imputation
print(df)

# Save the DataFrame to a CSV file
df.to_csv('filled_dataset.csv', index=False)

# Code to download the CSV file (works in Jupyter Notebook or Google Colab)
from IPython.display import FileLink

# Provide a link to download the CSV file
display(FileLink('filled_dataset.csv'))
```

## Visualizing the outliers

```python
sns.boxplot(data=df)
plt.show()
```

## Finding the outliers row numbers

```python
Q1 = df.quantile(.10)
Q3 = df.quantile(.90)
IQR = Q3 - Q1
outliers_iqr = ((df< (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_indices_iqr = outliers_iqr[outliers_iqr].index
```

```python
len(outliers_indices_iqr)
```

## Dropping outliers rows

```python
df1 = df.drop(outliers_indices_iqr,axis=0)
```

```python
sns.boxplot(data=df1)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

## Segregating the response features and target feature

```python
X = df1.drop('medv',axis=1)
y = df1['medv']
```

## Splitting the dataset into train data and test data

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

## Getting Started with model training using Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

## Showing the metrics of the model performance

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
adjusted_r2_score = 1-((1-r2_score(y_test,y_pred))/(df.shape[0]-df.shape[1]-1))
```

```python
adjusted_r2_score
```

## Showing plot b/w residual and Predicted Values

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

## Showing the regression line b/w true and predicted values

```python
plt.scatter(y_test, y_pred)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('True vs. Predicted Values')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.show()
```

## Applying CrossValidation using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = LinearRegression()
params = {
    'fit_intercept': [True, False],
    'n_jobs': [1, 2, 3,4],
    'copy_X': [True, False],
    'positive': [True,False]
}
grid_search = GridSearchCV(model, params, cv=5, scoring='accuracy')
grid_search.fit(X, y)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy Score:", grid_search.best_score_)
```

```python
model = LinearRegression(**grid_search.best_params_)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
r2_score(y_test,y_pred)
```
